```python
x=df_final.drop('Rent' , axis=1)
y=df_final['Rent']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.20 , random_state=42)

```

y_pred = model.predict(x_test)
